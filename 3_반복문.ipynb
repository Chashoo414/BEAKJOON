# 1. [2739번] 구구단

```python
N = int(input())

for i in range(1, 10) :
    result = N * i
    print('{} * {} = {}'.format(N,i,result))
```

# 3. [8393번] 합

```python
n = int(input())

total = 0

for i in range(1, n+1) :
    total += i

print(total)
```

# 4. [25304번] 영수증

```python
X = int(input())
N = int(input())

total = 0

for _ in range(N) :
    a, b = map(int, input().split())
    total += a * b

if X == total :
    print('Yes')
else :
    print('No')
```

# 10. [10871번] X보다 작은 수

```python
N, X = map(int, input().split())

A = list(map(int, input().split()))

for i in range(len(A)) :
    if A[i] < X :
        print(A[i], end=' ')
```

# 13. [1110번] 더하기 사이클

```python
# My solution
N = int(input())

cycle = 0
T = N

while True :
    if T >= 10 :
        ori_num = str(T)
    else :
        ori_num = '0' + str(T)

    add_num = str(int(ori_num[0]) + int(ori_num[-1]))

    new_num = int(ori_num[-1] + add_num[-1])

    cycle += 1
    T = new_num

    if new_num == N :
        break

print(cycle)
```

```python
# Solution based of short coding
N = int(input())

cycle = 0
T = N

while True :
    new_num = (T%10)*10 + T*11//10%10

    cycle += 1
    T = new_num

    if new_num == N :
        break

print(cycle)
```
