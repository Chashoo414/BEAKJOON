# 1. [2739번] 구구단

```python
N = int(input())

for i in range(1, 10) :
    result = N * i
    print('{} * {} = {}'.format(N,i,result))
```

# 2. [10950번] A+B - 3

```python
T = int(input())

for _ in range(T) :
    A, B = map(int,input().split())
    print(A+B)
```

# 3. [8393번] 합

```python
n = int(input())

total = 0

for i in range(1, n+1) :
    total += i

print(total)
```

# 4. [25304번] 영수증

```python
X = int(input())
N = int(input())

total = 0

for _ in range(N) :
    a, b = map(int, input().split())
    total += a * b

if X == total :
    print('Yes')
else :
    print('No')
```
