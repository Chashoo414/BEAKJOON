# 7. [10869번] 사칙연산

```python
A, B = map(int, input().split())
print(A+B, A-B, A*B, A/B, A%B, sep='\n')
```

## solution

> 파이썬의 print()함수에는 다양한 옵션들이 있다.<br>

> - `sep` : separation의 약자로, 출력하는 문자 사이에 구분자를 지정할 수 있다.
    
    print('Apple', 'Peach', 'Pear', sep='@')
    출력 >>> Apple@Peach@Pear
> - `end` : `end=""` 옵션은 그 뒤의 출력값을 이어서 출력한다.
    - sep처럼 구분자를 지정하여 사용할 수 있다.

    print('Apple', end='')
    print('Peach')
    출력 >>> ApplePeach

    print('Apple', end=' and ')
    print('Peach')
    출력 >>> Apple and Peach
> - `format` : 지정한 형식으로 문자를 출력한다.
    
    print('오늘은 {}월 {}일 입니다.'.format(9, 23))

# 10. [3003번] 킹, 퀸, 룩, 비숍, 나이트, 폰

# 11. [10430번] 나머지

```python
A, B, C = map(int, input().split())

result1 = (A+B)%C
result2 = ((A%C) + (B%C))%C
result3 = (A*B)%C
result4 = ((A%C) * (B%C))%C

print(result1, result2, result3, result4, sep='\n')
```

# 12. [2588번] 곱셈

```python
a = int(input())
b = input()

re1 = a * int(b[-1])
re2 = a * int(b[1])
re3 = a * int(b[0])
re4 = re1 + re2 * 10 + re3 * 100

print(re1, re2, re3, re4, sep='\n')
```
