# 1. [1330번] 두 수 비교하기

```python
A, B = map(int, input().split())

if A > B :
    print('>')
elif A < B :
    print('<')
else :
    print('==')
```

# 2. [9498번] 시험 성적

```python
score = int(input())

if score >= 90 :
    print('A')
elif score >= 80 :
    print('B')
elif score >= 70 :
    print('C')
elif score >= 60 :
    print('D')
else :
    print('F')
```

# 3. [2753번] 윤년

```python
year = int(input())

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0) :
    print(1)
else :
    print(0)
```

# 4. [14681번] 사분면 고르기

```python
x = int(input())
y = int(input())

if x > 0 and y > 0 :
    print(1)
elif x > 0 :
    print(4)
elif y > 0 :
    print(2)
else :
    print(3)
```

# 5. [2884번] 알람 시계

```python
H, M = map(int, input().split())    #1

h, m = H, M-45  #2

if m < 0 :      #3
    m += 60
    h -= 1

    if h < 0 :
        h += 24

print(h, m)     #4
```

# 6. [2525번] 오븐 시계

```python
h, m = map(int, input().split())
cook = int(input())

m += cook

while m >= 60 :
    m -= 60
    h += 1
    if h >= 24 :
        h -= 24

print(h, m)
```

# 7. [2480번] 주사위 세개

```python
a, b, c = map(int,input().split())
number = [a, b, c]

if a == b :
    if a == c :
        price = 10000 + a * 1000
    else :
        price = 1000 + a * 100

else :
    if a == c or b == c :
        price = 1000 + c * 100
    else :
        price = max(number) * 100
    
print(price)
```
