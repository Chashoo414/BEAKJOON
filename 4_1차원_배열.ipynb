# 4. [2562번] 최댓값

```python
num_list = []

for _ in range(9) :
    n = int(input())
    num_list.append(n)

m = max(num_list)

print(m)
print(num_list.index(m)+1)      
```

# 5. [5597번] 과제 안 내신 분..?

```python
student_list = [x for x in range(1, 31)]

for _ in range(28) :
    n = int(input())
    student_list.remove(n)

print(min(student_list))
print(max(student_list))
```

# 6. [3052번] 나머지

```python
remainder = {}

for _ in range(10) :
    n = int(input()) % 42

    if n in remainder :
        remainder[n] += 1
    else :
        remainder[n] = 1

print(len(remainder))
```

# 7. [1546번] 평균

```python
N = int(input())

score = list(map(int, input().split()))

m = max(score)

new_score = [x/m*100 for x in score]

mean = sum(new_score)/len(new_score)
print(f'{mean:0.2f}')
```

# 8. [8958번] OX퀴즈

```python
T = int(input())

for _ in range(T) :
    quiz = input()

    total = 0
    score = 1

    for i in range(len(quiz)) :
        if quiz[i] == 'O' :
            total += score
            score += 1
        else :
            score = 1
    
    print(total)
```

# 9. [4344번] 평균은 넘겠지

```python
C = int(input())

for _ in range(C) :
    score = list(map(int, input().split()))

    N = score.pop(0)

    average = sum(score) / N

    AAS = [x for x in score if x > average]        # AAS: Above Average Students

    AAS_rate = (len(AAS) / len(score)) * 100

    print(f'{AAS_rate:0.3f}%')
```
