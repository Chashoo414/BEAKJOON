# 2. [2562번] 최댓값

```python
num_list = []

for _ in range(9) :
    n = int(input())
    num_list.append(n)

m = max(num_list)

print(m)
print(num_list.index(m)+1)      
```

# 3. [3052번] 나머지

```python
remainder = {}

for _ in range(10) :
    n = int(input()) % 42

    if n in remainder :
        remainder[n] += 1
    else :
        remainder[n] = 1

print(len(remainder))
```

# 4. [1546번] 평균
